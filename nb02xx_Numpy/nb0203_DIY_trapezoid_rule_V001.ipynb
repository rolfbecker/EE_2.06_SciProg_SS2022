# DIY Trapezoid Rule

## Plot the function!

```python
import numpy as np
#np.arange
#np.array
#np.linspace
```

```python
def f(z):
    return z**3 - 20*z
```

```python
x = np.arange(-10,10.0001,0.1)
y = f(x)
```

```python
n = 5 # num of trapezoids
a = -4.
b = 0.
dx = (b-a)/n
print("width dx: ", dx)
```

```python
#help(np.linspace)
xs = np.linspace(a,b,n+1)
y0 = xs*0
ys = f(xs)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,1,figsize=(6,4),dpi=136)
ax.plot(x,y)
ax.plot(xs,y0,"o")
ax.plot(xs,ys,"o")
ax.set_xlim(-5,5)
ax.set_ylim(-50,50)
ax.grid(True)
plt.show()
```

## Trapezoid Rule

$A_i =h \cdot \Delta x =  {f(x_i) + f(x_{i+1}) \over 2} \cdot \Delta x$

$S = \Sigma_{i=0}^{n-1} A_i \approx \int_a^b f(x)dx $

$\lim_{n\rightarrow\infty} \Sigma_{i=0}^{n-1} A_i = \int_a^b f(x)dx $
