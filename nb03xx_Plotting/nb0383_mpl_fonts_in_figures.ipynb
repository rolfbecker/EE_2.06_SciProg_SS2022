```python
#mpl.rcParams['figure.dpi'] = 136
fig = plt.figure(figsize = (4,4),facecolor = "lightgrey", dpi = 136)
axes = fig.add_subplot(111)
axes.set_title("$f(x)=\int_0^x x'^2/2 dx'$")
axes.set_xlabel("This is the x-axis: $\int \ldots$")
axes.set_ylabel("This is the y-axis: $\int \ldots$")
axes.plot(range(11),"-o",color="orange") 
axes.axis('equal')
plt.show() 
```

```python
mpl.rcParams['figure.dpi'] = 136
plt.figure(figsize = (3,3),facecolor = "lightgrey")
plt.plot(range(11),"-o",color="orange") 
plt.title("$f(x)=\int_0^x x'^2/2 dx'$")
plt.xlabel("x-Label")
plt.ylabel("y-Label")
plt.show() 
```

```python
mpl.rcParams['figure.dpi'] = 136
fig = plt.figure(figsize = (3,3),facecolor = "lightgrey")
axes = fig.add_subplot(111)
axes.plot(range(11),"-o",color="orange") 
axes.set_title("$f(x)=\int_0^x x'^2/2 dx'$")
axes.set_xlabel("x-Label")
axes.set_ylabel("y-Label")
plt.show() 
```

```python
plt.rcParams["mathtext.fontset"] = "dejavuserif"
plt.figure(figsize=(12,10))
plt.plot(range(11),color="0.9")
plt.text(0.2,9,r"$\mathrm{The\ quick\ red\ dog\ jumped\ over\ the\ lazy\ brown\ fox}$",size=23)
plt.text(0.2,8,r"$\mathit{The\ quick\ red\ dog\ jumped\ over\ the\ lazy\ brown\ fox}$",size=23)
plt.show()
```
