```python
# I want to plot with matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt

# include the plots inline, i.e. in the text flow
%matplotlib inline
```

```python
# We need something to plot 
# numpy!
import numpy as np
```

```python
#pi = np.pi
#np.sin
from numpy import pi, sin, cos
x = np.arange(-4,4.1000001,0.01)*pi

ys = sin(x)
yc = cos(x)
```

```python
# Resize all figures. Default for the rest of the notebook
plt.rcParams['figure.figsize'] = (4,3) # Set the default figure size. You have to calibrate DPI
plt.rcParams['figure.dpi'] = 136 # depends on your system
plt.show()
```

```python
# generate a figure
fig1 = plt.figure()

# add an axis (coordinate system) to it.
# subplot(111) means use 1x1 plots and use number 1 of these plots.
ax1 = fig1.add_subplot(111)

# plot!
ax1.plot(x,ys,label=r"$\sin(x)$")
ax1.plot(x,yc,label=r"$\cos(x)$")
a

ax1.set_title(r'My Cool Plot 1')
ax1.set_xlabel(r'$x,\; [x] = \mathrm{rad}$')
ax1.set_ylabel(r'Amplitude')
ax1.legend(title="What you see:", loc = "upper right")
plt.show()
```
