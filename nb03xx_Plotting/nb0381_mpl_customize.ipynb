Matplotlib is probably the de facto standard library for Python plotting. However, other more powerful packages are available such as plotly or bokeh, but they are more difficult to use. 

The example below shows one of several methods how you can customize Matplotlib plots in Jupyter Notebook. All customisations are limited to this notebook, i.e. the scope of the settings is local to the notebook.

In case you want to change the default plotting options for all your plot you might consider to do that in the config file named `matplotlibrc` in a hidden directory (folder) in your personal home directory, e.g. `C:\Users\johndoe\.matplotlib`.

These default settings can be overridden later in your individual Python scripts and Notebooks, of course.

## Let us customize our plots ##

```python
# Import the whole matplotlib package and matplotlib.pyplot explicitly. pyplot is used to generate the plots.
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### Choose your default figure size on your computer screen ###

```python
# Your intended figure size in cm x cm on the display / monitor
xcm = 15.0 ; ycm = 10.0  # default figure size in cm
dpi = 136 # you have to tune this value until the real figure size in inch x inch mathes the intended size.

# Default figure size in inch x inch
xinch = xcm/2.54 ; yinch = ycm/2.54  # in the rc settings it has to be inch.

print("figsize in cm  : %.1f x % .1f" % (xcm, ycm))
print("figsize in inch: %.1f x % .1f" % (xinch, yinch))

mpl.rcParams['figure.figsize'] = (xinch,yinch) # Set the default figure size. You have to calibrate DPI
mpl.rcParams['figure.dpi'] = dpi

plt.show()
```

### Generate the example data to be plotted. ### 

$$f(x) = \frac{x^3}{50}-\frac{3}{2}x \qquad \mathrm{for}\, x \in \{-12,-11.9,\ldots,0,\ldots,+11.9,+12\}$$

```python
import numpy as np
```

```python
x = np.arange(-12.,12.01,1)
y = x**3/50. - 3./2. * x
```

```python
fig = plt.figure(1)
axes = fig.add_subplot(111)
axes.set_title("$f(x)=x^3 / 50 -3/2 x + \int_{-\infty}^{+\infty}x^3 dx$")
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.plot(x, y,"-o",color="orange") # local definitions: line and marker style, color
plt.show() # This makes Jupyter read the resources!
```

### Some printing customization ###

```python
mpl.rcParams['figure.facecolor'] = 'lightgrey'
mpl.rcParams['font.size'] = 10 # basic font size. Title, legend, label and other font sizes are scaled relatively to it.
mpl.rcParams['lines.linewidth'] = 3
mpl.rcParams['axes.grid'] = True

# So many different fonts ...

# To use DejaVu Serif set:
mpl.rcParams['font.serif'] = 'DejaVu Serif'
mpl.rcParams["mathtext.fontset"] = "dejavuserif"
mpl.rcParams['font.family'] = "serif"

# The DEFAULT in Matplotlib 2.0 is font DejaVu Sans:
#mpl.rcParams['font.sans-serif'] = 'DejaVu Sans'
#mpl.rcParams['mathtext.fontset'] = 'dejavusans'
#mpl.rcParams['font.family'] = 'sans-serif'
# mpl.rcParams['font.size'] = 10

# THIS LOOKS MORE LIKE MATH TEXT BOOK! Esp. integrals are better readable.
#mpl.rcParams['font.serif'] = 'STIXGeneral'
#mpl.rcParams["mathtext.fontset"] = "stix"
#mpl.rcParams['font.family'] = "serif"
```

### Figure 1 ###
  * Default definition from above
  * Local definitions: 
    * line style 
    * marker style
    * line color

```python
fig = plt.figure(1)
axes = fig.add_subplot(111)
axes.set_title("$f(x)=x^3 / 50 -3/2 x + \int_{-\infty}^{+\infty}x^3 dx$")
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.plot(x, y,"-o",color="orange") # local definitions: line and marker style, color
plt.show() # This makes Jupyter read the resources!
```

### Figure 2 ###
  * Default definition from above
  * Local definitions:
    * figure size in inch x inch 
    * line style 
    * marker style
    * marker face color
    * marker size
    * tilte font size

```python
fig = plt.figure(2,figsize=(4,3))
axes = fig.add_subplot(111)

# All definitions in the plot command are just valid in the scope of the current axes object.
axes.plot(x, y,color="red", linestyle='dashed',  linewidth=3, marker='o',
         markerfacecolor='blue', markersize=5, label= '$f_1(x)= x^3 / 50 -3/2 x$')
axes.plot(x, -y,color="green", linewidth=5, label= '$f_2 = -f_1$')

axes.legend() # shows legend on plotting area.
axes.set_title("$f(x)=x^3 / 50 -3/2 x + \int_{-\infty}^{+\infty}x^3 dx$", fontsize=20)
axes.grid(False)
axes.set_xlabel('x')
axes.set_ylabel('y')


plt.show()
```

### Figure 3: Use a Plot Style Sheet ###

```python
# If you call this function you will change the settings for all future plots in your Notebook.
#plt.style.use('ggplot')
```

```python
fig = plt.figure(3)

# This limits the scope of the plot style sheet to the current plot.
with plt.style.context(('ggplot')):
    axes = fig.add_subplot(111)
    axes.set_title("$f(x)=x^3 / 50 -3/2 x + \int_{-\infty}^{+\infty}x^3 dx$")
    axes.set_xlabel("x")
    axes.set_ylabel("y")
    axes.plot(x, y)
    
plt.show() # This makes Jupyter read the resources!
```

```python
fig = plt.figure(4)

mpl.rcParams['mathtext.fontset'] = 'cm'
mpl.rcParams['mathtext.rm'] = 'serif'

# This limits the scope of the plot style sheet to the current plot.
with plt.style.context(('classic')):
    axes = fig.add_subplot(111)
    axes.set_title("$f(x)=x^3 / 50 -3/2 x + \int_{-\infty}^{+\infty}x^3 dx$")
    axes.set_xlabel("x")
    axes.set_ylabel("y")
    axes.plot(x, y)
    
plt.show() # This makes Jupyter read the resources!
```

```python
#mpl.rcParams['mathtext.fontset'] = 'cm'
#mpl.rcParams['mathtext.rm'] = 'serif'

#mpl.rcParams['font.serif'] = 'DejaVu Serif'
mpl.rcParams['font.serif'] = 'STIXGeneral'
mpl.rcParams["mathtext.fontset"] = "stix"
mpl.rcParams['font.family'] = "serif"

fig = plt.figure(5)

axes = fig.add_subplot(111)
axes.set_title(r"$f(x)=x^3 / 50 -3/2 x + \int_0^{+\infty}\frac{x^3}{1}dx$")
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.plot(x, y)
    
plt.show() # This makes Jupyter read the resources!
```
