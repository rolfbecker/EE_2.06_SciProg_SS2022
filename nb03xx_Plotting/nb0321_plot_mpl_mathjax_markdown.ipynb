# Plotting with Matplotlib #

```python
# I want to plot with matplotlib
%matplotlib inline

#import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np
```

```python
x = np.arange(-4,4.0000001,0.01)
```

```python
from numpy import pi, sin, cos
x = np.arange(-4,4.0000001,0.01)*pi

ys = sin(x)
yc = cos(x)
```

```python
# Resize all figures. Default for the rest of the notebook
plt.rcParams['figure.figsize'] = (4,3) # Set the default figure size. You have to calibrate DPI
plt.rcParams['figure.dpi'] = 136 # depends on your system
```

```python
# generate a figure
# use default figsize ...
fig1 = plt.figure()

# ... or override locally for this very figure.
#fig1 = plt.figure(figsize=(4,3), dpi = 136)

# add an axis (coordinate system) to it.
# subplot(111) means use 1x1 plots and use number 1 of these plots.
ax1 = fig1.add_subplot(111)

# plot!
ax1.plot(x,ys,"-",label=r"$\sin(x)$")
ax1.plot(x,yc,"-",label=r"$\cos(x)$")

ax1.legend(title="What you see:", loc = "lower left")

ax1.set_title(r'My Cool Plot 1')
ax1.set_xlabel(r'$x,\; [x] = \mathrm{rad}$')
ax1.set_ylabel(r'Amplitude')

ax1.grid(True)

plt.show()
```

## Exercise ##

A functions is given:

$f(x) = x e^{-x^2}$

Plot $f(x)$ for $x \in [0,10]$ or $0 \le x \le 10$

i.e. in the limits of 0 to 10.

```python
from numpy import exp
```

```python
def f(x):
    return x*exp(-x**2)
```

```python
# don't us list, 
# use np.arange() because it allows vectorization
x = np.arange(0,10.000001,0.01)
y = f(x)
```

```python
xs = np.arange(0,10.00001,0.5)
ys = f(xs)
y0 = np.zeros(xs.shape)
```

```python
fig2 = plt.figure(figsize=(4,3),dpi = 136)
ax2 = fig2.add_subplot(111)
hand1 = ax2.plot(x,y,label="function")
hand2 = ax2.plot(xs,ys,"*-",label="samples")
hand3 = ax2.plot(xs,y0,"x")
ax2.set_xlim(0,3)
ax2.set_title(r"$f(x) = xe^{-x^2}$")
ax2.set_xlabel("$x$")
ax2.set_ylabel("$f(x)$")
ax2.legend(loc = "upper right")
ax2.grid(True)
plt.show()
```
