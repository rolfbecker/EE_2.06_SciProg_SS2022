```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
#url = "https://www.cndc.org.ni/siimem/graficos/generacion%20sin/generacion.php"
url = "https://www.cndc.org.ni/siimem/graficos/generacion%20sin/Generacion_consulta.php"
```

```python
r = requests.get(url) # response object
#r.content
```

```python
r.content
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python
#print('Classes of each table:')
#for table in soup.find_all('table'):
#    print(table.get('class'))
```

```python
td = soup.findAll('td')
print(td[100])
print(td[101])
print(td[102])
```

```python
td = soup.findAll('td')
print(td[100].find(text=True))
print(td[101].find(text=True))
print(td[102].find(text=True))
```

```python
rows = soup.findAll('tr')
for tr in rows:
    cols = tr.findAll('td')
    for td in cols:
        print(td.find(text=True))   
```

```python
data={}

rows = soup.findAll('tr')

for tr in rows:
    lst = [td.find(text=True) for td in tr.findAll('td')]
    print(lst)
    if len(lst)==2: data[lst[0]]=lst[1]
```

```python
data.keys()
```

```python
data["PEN3"]
```

```python
geo_gen = ["PEN3","PEN4","PMT1", "PMT2", "PMT3"]
```

```python
for key in geo_gen:
    print(f"{key:4s} : {float(data[key]):.2f} MW")
```
