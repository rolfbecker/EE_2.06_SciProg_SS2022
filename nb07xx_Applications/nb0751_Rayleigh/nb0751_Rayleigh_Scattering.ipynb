# Rayleigh Scattering

$$
I = I_0 \frac{ 1+\cos^2 \theta }{2 R^2} \left( \frac{ 2 \pi }{ \lambda } \right)^4 \left( \frac{ n^2-1}{ n^2+2 } \right)^2 \left( \frac{d}{2} \right)^6
$$

https://en.wikipedia.org/wiki/Rayleigh_scattering



$$
\begin{eqnarray}
I & = & \frac{I_0}{4\pi R^2} \left(1+\cos^2 \theta \right) 2\pi \left( \frac{ 2 \pi }{ \lambda } \right)^4 \left( \frac{ n^2-1}{ n^2+2 } \right)^2 \left( \frac{d}{2} \right)^6 \\
& = & \frac{I_0}{4\pi R^2} \left(1+\cos^2 \theta \right)  \frac{(2 \pi)^5}{\lambda^4} \frac{d^6}{2^6} \left( \frac{ n^2-1}{ n^2+2 } \right)^2\\
& = & \frac{I_0}{4\pi R^2} \left(1+\cos^2 \theta \right)  \pi^5 \frac{d^6}{\lambda^4} \frac{2^5}{2^6} \left( \frac{ n^2-1}{ n^2+2 } \right)^2\\
\end{eqnarray}
$$




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rayleigh(theta, I0 = 1, R = 1, lam = 2*np.pi, n = 1e3, d = 2):
    t1 = I0
    t2 = (1+np.cos(theta)**2)/(2*R**2)
    t3 = (2*np.pi/lam)**4
    t4 = ((n**2 - 1)/(n**2 + 2))**2
    t5 = (d/2)**6
    I = t1 * t2 * t3 *t4 * t5
    return I
```

```python
theta_deg = np.arange(0.,361, 5)
theta = theta_deg/360*2*np.pi
```

```python
# for d,r in zip(theta_deg, theta):
#    print(f"{d:5.1f} deg -> {r:5.3f} rad")  
```

## Rayleigh Scattering: Geometric properties and relative strength.

```python
r = rayleigh(theta)
```

```python
fig1, ax1 = plt.subplots(subplot_kw={'projection': 'polar'}, figsize = (5,5), dpi = 136)
ax1.plot(theta, r, "b.-", label = "relative power density")
ax1.legend()
print("Consider the scale of the radius.")
plt.show()
```

## Rayleigh Scattering for 400 nm and 700 nm

```python
lam1 = 400.e-9 # 400 nm, blue
lam2 = 700.e-9 # 700 nm, red

n = 2.0
d = 10.0e-10 # 10 Angström 
R = 1.0e3 # 1 km
```

```python
r1 = rayleigh(theta, d = d, lam = lam1, R = R, n = n)
r2 = rayleigh(theta, d = d, lam = lam2, R = R, n = n)
```

```python
fig2, ax2 = plt.subplots(subplot_kw={'projection': 'polar'}, figsize = (5,5), dpi = 136)
ax2.plot(theta, r1, "b-", label = "400nm")
ax2.plot(theta, r2, "r-" , label = "700nm")
ax2.legend(title = "scattered power density")
ax2.set_title("Rayleigh Scattering")
# ax2.set_xlabel("scatter angle")
# ax2.set_ylabel("scattered power density")
print("Consider the scale of the radius, \ni.e. the radiant power density \nscattered by a single molecule.")
plt.show()
```
